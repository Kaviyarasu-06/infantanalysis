```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
print(numeric_df.columns)
```

```python
skewness = numeric_df.skew()
display(skewness)



plt.figure(figsize=(20,15))
plotnumber=1
for column in numeric_df:
    ax=plt.subplot(6,6,plotnumber)
    sns.kdeplot(numeric_df[column])
    plt.xlabel(column)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

# skewness is high for only jaundice_level

```python
df['jaundice_level_log'] = np.log1p(df['jaundice_level_mg_dl'])
display(df[['jaundice_level_mg_dl', 'jaundice_level_log']].head())
```

```python
skewness_jaundice_log = df['jaundice_level_log'].skew()
print(f"Skewness of 'jaundice_level_log': {skewness_jaundice_log}")
```

```python
correlation_matrix = numeric_df.corr()
display(correlation_matrix.head())
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
print(df['apgar_score'].value_counts())
```

# now there is no null values

```python
apgar_mode = df['apgar_score'].mode()[0]
df['apgar_score'].fillna(apgar_mode, inplace=True)
df['apgar_score']
```

# week column from the date

```python
df['date'] = pd.to_datetime(df['date'])
df['week_of_year'] = df['date'].dt.isocalendar().week
display(df[['date', 'week_of_year']].head(10))
```

```python
df['age_weeks'] = df['age_days'] / 7
print(df['age_weeks'])
```

```python
df['length_m'] = df['length_cm'] / 100
df['bmi'] = df['weight_kg'] / (df['length_m'] ** 2)
display(df[['weight_kg', 'length_cm', 'length_m', 'bmi']].head())
```

# daily changing weight of babies

```python
df_encoded = df.sort_values(by=['baby_id', 'date'])
df_encoded['previous_day_weight_kg'] = df.groupby('baby_id')['weight_kg'].shift(1)
df_encoded['weight_change_from_previous_day'] = df['weight_kg'] - df_encoded['previous_day_weight_kg']

display(df_encoded[['baby_id', 'date', 'weight_kg', 'previous_day_weight_kg', 'weight_change_from_previous_day']].head(10))
```

```python
numerical_cols = ['gestational_age_weeks', 'birth_weight_kg', 'weight_kg', 'temperature_c', 'heart_rate_bpm']

for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df_encoded[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
average_birth_metrics = df[['birth_weight_kg', 'birth_length_cm']].mean()

plt.figure(figsize=(8, 5))
average_birth_metrics.plot(kind='bar')
plt.title('Average Birth Weight and Birth Length')
plt.ylabel('Measurement')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_encoded['oxygen_saturation'], kde=True)
plt.title('Distribution of Oxygen Saturation')
plt.xlabel('Oxygen Saturation')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_encoded, x='gestational_age_weeks', y='birth_weight_kg')
plt.title('Relationship between Gestational Age and Birth Weight')
plt.xlabel('Gestational Age (weeks)')
plt.ylabel('Birth Weight (kg)')
plt.show()
```
